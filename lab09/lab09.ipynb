# 绘制切线

**求解切线方程**

```python
# y = f(x) = x^2 - e(-x)
# f'(x) = 2*x + e(-x)
# 令 x=1，求解斜率k=2
#求解点坐标(1,1-e(-1))
#用点斜式写出切线方程 y = (2+e(-1))(x-1)+1-e(-1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 3, 100)
y1 = x**2 - np.exp(-x)
y3 = (2+np.exp(-1))*(x-1)+1-np.exp(-1)

plt.plot(x, y1, label='y=x^2-e^(-x)')
plt.plot(x, y3, label='y=(2+e^(-1))(x-1)+1-e^(-1)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
x = d2l.arange(0, 3, 0.01)
y1 = x**2 - np.exp(-x)
y2 = (2+np.exp(-1))*(x-1)+1-np.exp(-1)
d2l.plt.figure(figsize=(6, 4))
d2l.plt.plot(x, y1)
d2l.plt.plot(x, y2, 'purple', ls='--')
d2l.plt.xlabel('x')
d2l.plt.ylabel('f(x)')
d2l.plt.title('Tangent Line')
d2l.plt.legend(['f(x)', 'Tangent Line (x=1)'])
d2l.plt.grid(alpha=0.3)
d2l.plt.show()
```
